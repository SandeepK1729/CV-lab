```python
from matplotlib import pyplot as plt
```

```python
ages = [1, 2, 3, 2, 4, 2, 3, 5, 3, 1, 2, 7, 6, 6, 6, 6, 7, 8, 9]
range = (min(ages), max(ages))
bins = len(set(ages))
```

```python
plt.hist(ages, bins, range, color = 'blue', histtype = 'bar', rwidth = 0.8)
plt.title("Histograme")
plt.xlabel("ages")
plt.ylabel("frequency")
plt.show()
```
