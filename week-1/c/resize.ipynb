```python
import cv2
```

```python
path = "../../images/"

file_names = ["cloudy.jpg", "sky.jpg", "light.jpg"]
```

```python
for file_name in file_names : 
    img = cv2.imread(path + file_name)
    
    img = cv2.resize(img, (300, 300))
    
    cv2.imshow(file_name, img)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
