### 1.b reading and writing of video

```python
import cv2
```

```python
vid = cv2.VideoCapture('videos/earth.mp4')
dim = tuple(map(lambda x : int(x / 1), [vid.get(3), vid.get(4)]))
fps = int(vid.get(5))
cap = cv2.VideoWriter('videos/copy of earth.mp4', cv2.VideoWriter_fourcc(*"mp4v"), fps, dim)
```

```python
while vid.isOpened():
    ret, frame = vid.read()
    
    if not ret:
        break
        
    img = cv2.cvtColor(cv2.resize(frame, dim), cv2.COLOR_BGR2RGB)
    cap.write(img)
    
cap.release()
vid.release()
```

### 1c : image enhancement

```python
from PIL import Image, ImageEnhance
```

```python
img = Image.open('images/doremon.jpg')
```

```python
enhancers = {
    'Brightness' : ImageEnhance.Brightness(img),
    'Sharpness' : ImageEnhance.Sharpness(img),
    'Color' : ImageEnhance.Color(img),
    'Contrast' : ImageEnhance.Contrast(img),
}
```

```python
def Enhancer(enhancer):
    plt.figure(figsize = (20, 50))
    
    for i in range(1, 6):
        plt.subplot(1, 5, i)
        plt.imshow(enhancer.enhance(i))

    plt.show()
```

```python
for name, enhancer in enhancers.items():
    Enhancer(enhancer)
```

### 4 different filtering operations

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img          = cv2.imread('images/doremon.jpg')
dim = (257, 300)
img          = cv2.resize(img, dim)
gray_img     = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
images = {
    'original image' : img,
    'gray image' :  gray_img,
    'Gaussian Blur' : cv2.GaussianBlur(img, (3, 3), 0),
    'medium Blur' : cv2.medianBlur(img, 3),
    'bilateral blur' : cv2.bilateralFilter(img, 15, 10, 100)
}
```

```python
def show(images):
    for title, image in images.items():
        cv2.imshow(title, image)
        cv2.waitKey(0)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

### 5 Morphological Operations

```python
import cv2
import numpy as np
```

```python
img   = cv2.resize(cv2.imread('images/nebula.jpg', 0), (512, 288))
mask  = np.ones((5, 5))
```

```python
images = {
    'original image' : img,
    'dilated image'  : cv2.dilate(img, mask),
    'eroded image'   : cv2.erode(img, mask),
    'opening'        : cv2.morphologyEx(img, cv2.MORPH_OPEN, mask),
    'closing'        : cv2.morphologyEx(img, cv2.MORPH_CLOSE, mask),
    
}
```

```python
show(images)
```
