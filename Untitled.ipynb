```python
import numpy as np
import cv2
```

```python
img = cv2.resize(
            cv2.imread('images/diamond.png'),
            (300, 300)
        )

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
sift_obj = cv2.SIFT_create()
kp = sift_obj.detect(gray, None)
```

```python
img2 = cv2.resize(
            cv2.drawKeypoints(gray, kp, img),
            (300, 300)
        )
```

```python
cv2.imshow("Output", img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
