```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(
        '../images/beach.jpg', 
        0
    )
```

```python
_, mask = cv2.threshold(
        img, 
        210, 
        255, 
        cv2.THRESH_BINARY_INV
    )
```

```python
images = [
    ('Original image', img), 
    ('masked image'  , mask)
]
```

```python
for i in range(len(images)):
    title, image = images[i]
    
    plt.subplot(1, 2, i + 1)
    plt.imshow(image)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
plt.show()
```
