```python
import cv2
import numpy as np
```

```python
img = cv2.cvtColor(
        cv2.imread('../images/beach.jpg'), 
        cv2.COLOR_BGR2GRAY
    )
```

```python
thresholding_techniques = [
    cv2.THRESH_BINARY, 
    cv2.THRESH_BINARY_INV,
    cv2.THRESH_TRUNC, 
    cv2.THRESH_TOZERO, 
    cv2.THRESH_TOZERO_INV
]
```

```python
titles = [
    'Binary Threshold', 
    'Binary Threshold inverted', 
    'Tuncated Threshold', 
    'set to 0', 
    'set to zero invetred'
]
```

```python
thresh_images = []
for threshold_technique in thresholding_techniques:
    _, image = cv2.threshold(
            img, 
            120, 
            255,
            threshold_technique
        ) 
    thresh_images.append(image)
```

```python
for i in range(5): 
    cv2.imshow(
        titles[i], 
        cv2.resize(thresh_images[i], (300, 300))
    )
    cv2.waitKey(0)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
