```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from math import ceil
```

```python
img = cv2.imread( '../images/sunset.jpg', 0)
_, mask = cv2.threshold(img, 210, 255, cv2.THRESH_BINARY_INV )
```

```python
kernal = np.ones((2, 2), np.uint8)
```

```python
images = [
    ('original image',  img), 
    ('masked image'  ,  mask), 
    ('dilated image' ,  cv2.dilate(mask, kernal, iterations = 2  )), 
    ('erosion image' ,  cv2.erode( mask, kernal, iterations = 1  )), 
    ('opening image' ,  cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)), 
    ('closing image' ,  cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)), 
    #('gradient image',  cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernal)),
    #('top hat image' ,  cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernal)), 
    #('black hat image', cv2.morphologyEx(mask, cv2.MORPH_BLACKHAT, kernal))
]
```

```python
n = len(images)
for i in range(n):
    title, image = images[i]
    
    plt.subplot(ceil(n / 3), 3, i + 1)
    plt.imshow(image)
    #plt.figure(figsize = (2, 2))
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
plt.show()
```
